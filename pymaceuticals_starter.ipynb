```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_drug_clinical = pd.merge(clinical_trial_data, mouse_drug_data, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
combined_drug_clinical.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor =combined_drug_clinical.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
avg_tumor = pd.DataFrame(avg_tumor).reset_index()
avg_tumor
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor =combined_drug_clinical.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame and preview DataFrame
sem_tumor = pd.DataFrame(sem_tumor).reset_index()
sem_tumor
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_tumor_pivot = avg_tumor.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
avg_tumor_pivot
```

```python
# Minor Data sem munging
sem_tumor_pivot = sem_tumor.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
sem_tumor_pivot
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(avg_tumor_pivot.index, avg_tumor_pivot["Capomulin"], 
             yerr=sem_tumor_pivot["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_tumor_pivot.index, avg_tumor_pivot["Infubinol"], 
             yerr=sem_tumor_pivot["Infubinol"], color="b", marker="*", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_tumor_pivot.index, avg_tumor_pivot["Ketapril"], 
             yerr=sem_tumor_pivot["Ketapril"], color="g", marker="x", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_tumor_pivot.index, avg_tumor_pivot["Placebo"], 
             yerr=sem_tumor_pivot["Placebo"], color="y", marker="+", markersize=5, linestyle="dashed", 
             linewidth=0.50)

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin','Ketapril','Infubinol','Placebo'],loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("analysis1.png")

#Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Minor Data Munging to Re-Format the Data Frames
sem_met_site_pivot = sem_met_site.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
sem_met_site_pivot
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(avg_met_site_pivot.index, avg_met_site_pivot["Capomulin"], 
             yerr=sem_met_site_pivot["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_met_site_pivot.index, avg_met_site_pivot["Ketapril"], 
             yerr=sem_met_site_pivot["Ketapril"], color="b", marker="*", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_met_site_pivot.index, avg_met_site_pivot["Infubinol"], 
             yerr=sem_met_site_pivot["Infubinol"], color="g", marker="x", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.errorbar(avg_met_site_pivot.index, avg_met_site_pivot["Placebo"], 
             yerr=sem_met_site_pivot["Placebo"], color="y", marker="+", markersize=5, linestyle="dashed", 
             linewidth=0.50)

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met.Sites")
plt.xlabel("Treatment Duration (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin','Ketapril','Infubinol','Placebo'],loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("analysis2.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival=combined_drug_clinical.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
survival = pd.DataFrame({"Mouse Count":survival})
survival=survival.reset_index()

# Preview DataFrame
survival
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival=survival.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
survival
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * survival["Capomulin"] / 25, color="r", marker="o",  markersize=5, linestyle="dashed",
             linewidth=0.50)
plt.plot(100 * survival["Ketapril"] / 25, color="b", marker="*", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.plot(100 * survival["Infubinol"] / 25, color="g", marker="x", markersize=5, linestyle="dashed", 
             linewidth=0.50)
plt.plot(100 * survival["Placebo"] / 25, color="y", marker="+", markersize=5, linestyle="dashed", 
             linewidth=0.50)

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(['Capomulin','Ketapril','Infubinol','Placebo'],loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("analysis3.png")

# Show the Figure
plt.show()

```
